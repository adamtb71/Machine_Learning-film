```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import os
import csv
```

```python
movies_df = pd.read_csv('Movie_budget_and_revenue_v2.csv')
movies_df.head()
```

```python
# Dependencies
import requests

url = "http://www.omdbapi.com/?apikey=trilogy&t="
movies = movies_df['title'].values.tolist()  
```

```python
# solve rotten tomatoes problem. 
def get_tomato(movie):
    for x in movie['Ratings']:
        if x['Source'] =='Rotten Tomatoes':
            return x['Value']
```

```python
movie_name = []
plot =[]
metascore =[]
imdb_rating = []
rotten_tomatoes = []
i=1

for movie in movies:
    try:
        movie_data = requests.get(url + movie).json()
        movie_name.append(movie_data['Title'])
        plot.append(movie_data['Plot'])
        rotten_tomatoes.append(get_tomato(movie_data))
        metascore.append(movie_data['Metascore'])
        imdb_rating.append(movie_data['imdbRating'])
        print(f"Processsing record {i} | {movie}" )
        i=i+1
    except Exception:
        print("error")
```

```python
df_rotten_movie = pd.DataFrame(list(zip(movie_name,rotten_tomatoes)), columns = ['movie','rotten tomatoes'])
```

```python
#make a data frame for imdb ratings
df_imdb_movie = pd.DataFrame(list(zip(movie_name,imdb_rating)), columns = ['movie','imdb rating'])
```

```python
#make one for meta score
df_meta_movie = pd.DataFrame(list(zip(movie_name,metascore)), columns = ['movie','metascore'])
```
